```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Load image
img = Image.open("/content/Flower_stock_photo.jpg")

# Image-to-image transformation
translated_img = ImageOps.grayscale(img)

# Save output
translated_img.save("translated_output.png")

# Display images
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Input Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(translated_img, cmap="gray")
plt.title("Translated Image")
plt.axis("off")

plt.show()
```
